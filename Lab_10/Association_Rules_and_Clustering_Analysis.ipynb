```python
import pandas as pd
import numpy as np
df  = pd.read_csv('countries_continents.csv')
```

```python

for i in range(3,8,2):
        
    # Initialize KMeans with k clusters
    kmeans = KMeans(n_clusters=i, random_state=42)
    # Fit the model
    kmeans.fit(X)
    # Predict the cluster labels
    y_kmeans = kmeans.predict(X)
    
    print(y_kmeans)
    # Get the cluster centers
    centroids = kmeans.cluster_centers_
    # Plot the data points and centroids
    plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5)
    plt.show()
```

Part B: Optimal Cluster Determination
(a) Implement K-Means clustering for k ranging from 2 to 15
(b) For each k, calculate:
• WCSS (Within-Cluster Sum of Squares)
• Silhouette Score
(c) Create a visualization showing both metrics across all k values
(d) Identify the optimal number of clusters based on:
• Elbow point in WCSS curve
• Maximum Silhouette Score
Deliverable: One plot showing WCSS and Silhouette Score trends; Report the
optimal number of clusters with justification (2-3 sentences)

```python
from sklearn.metrics import silhouette_score
wcss = [] 
sil = []
for i in range(2,16):
        
    # Initialize KMeans with k clusters
    kmeans = KMeans(n_clusters=i, random_state=42)
    # Fit the model
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    sil.append(silhouette_score(X , kmeans.labels_))
    
```

```python

# Plot WCSS for different values of k
plt.plot(range(2, 16), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python

# Plot WCSS for different values of k
plt.plot(range(2, 16), sil)
plt.title('Silhouette Score Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

TASK 1: Frequent Rule on Filtered Transactions (Item Size = 4) [ 1.5 ]
(1) After performing the data cleaning steps, create a transactional basket (Invoice
× Items)(using crosstab).
(2) Filter the basket to include only transactions that contain exactly 4 items.
(3) Apply the fpgrowth algorithm to find frequent itemsets with a minimum sup-
port of 0.01.
(4) Generate association rules from these frequent itemsets using a minimum con-
fidence of 0.1
(5) Calculate the union size for each rule (number of items in antecedent + conse-
quent) and retain only those rules whose union size equals 4
(6) Identify and print the best rule sorted by confidence, support, and lift

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
basket = pd.crosstab(retail["InvoiceNo"] ,retail["Description"])
```

```python
basket
```

```python
basket = basket[basket.sum(axis=1) == 4]
```

```python
basket
```

```python
frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules
```

```python
rules['union_size'] = rules['antecedents'].apply(lambda x: len(x)) + rules['consequents'].apply(lambda x: len(x))

rules_filtered = rules[rules['union_size'] == 4]
rules_filtered
```

```python
best_rule = rules_filtered.sort_values(by=['confidence', 'support', 'lift'], ascending=False)

print("Best Association Rule (Union Size = 4):")
print(best_rule)
```

```python
basket = pd.crosstab(retail["InvoiceNo"] ,retail["Description"])
frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
```

```python
best_rule = rules.sort_values(by=['confidence', 'support', 'lift'], ascending=False).iloc[0]

print("Best Association Rule:")
print(best_rule['antecedents'])
print(best_rule)
```
