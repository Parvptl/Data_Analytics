```
def transpose(matrix):
    transposed_matrix = [[matrix[j][i] for j in range(len(matrix[0]))] for i in range(len(matrix))]
    print(transposed_matrix)

matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]
transpose(matrix)
```

## Question 20

```
import random
from collections import defaultdict

def freq():
    ls = []
    most_freq = 0
    dic = defaultdict(int)
    for i in range(100):
        ls.append(random.randint(1,200))
        dic[ls[-1]]+=1
        if(dic[ls[-1]]>dic[most_freq]):
            most_freq = ls[-1]
    print(dic)
    print("most frequent element is :", most_freq)
freq()
```

```
Q11)Write a function to find the largest gap between consecutive prime numbers in a given range [a, b].
```

## Question 22

```
def is_prime(n):

    if n==1 :
        return False
    if n==2 :
        return True
    
    i= 2 ;
    
    while(i*i<=n):
        if(n%i==0):
            return False
        i+=1
    
    return True

def max_gap(a,b):

    prev = -1
    max_gap = 0 
    for i in range(a,b+1):

        if(is_prime(i)):
            if(prev==-1):
                prev = i
            else:
                gap = i-prev
                prev=i
                if(gap>max_gap):
                    max_gap = gap

    return max_gap

a = int(input())
b = int(input())

print(max_gap(a,b))
        
```

## Question 24

```
def find_digital_root(n):

    num = 0 
    temp = n
    while(len(str(n))>1):
        num += temp%10 
        temp = temp//10
        if(temp==0):
            temp=num
            n = num 
            num = 0 

    return n

n = int(input())

print(find_digital_root(n))
```

## Question 26

```
def is_pattern(string):

    pattern1 = True
    pattern2 = True
    pattern3 = True
    pattern4 = True

    vowel = ['a' , 'e' , 'i' , 'o' , 'u']
    if(string[0] in vowel):
        pattern2 = False
    else:
        pattern1 = False

    if(len(string)<2):
        pattern3 = pattern4 = False
        
    for i in range(len(string)):

        if(i!=0 and pattern1):

            if(string[i] in vowel):
                if(string[i-1] in vowel):
                    pattern1 = False

            else:
                if(not string[i-1] in vowel):
                    pattern1 = False

        if(i!=0 and pattern2):

            if(string[i] in vowel):
                if(string[i-1] in vowel):
                    pattern1 = False

            else:
                if(not string[i-1] in vowel):
                    pattern1 = False
            
        if(pattern3):
            if(not string[i] in vowel):
                pattern3 = False

        if(pattern4):
            if(string[i] in vowel):
                pattern4 = False


    return pattern1 or pattern2 or pattern3 or pattern4

String = input()

if(is_pattern(String)):
    print("YES")

else:
    print("NO")
        
```

## Question 28

```
def count_fib(ls , fib) :
    count = 0 
    for i in ls :
        if i in fib:
            count+=1
    print(count)

fib = [0]*1001
fib[1] = 1

for i in range(2,1001):
    fib[i] = fib[i-1] + fib[i-2]


ls=[int(x) for x in input().split()]



count_fib(ls , fib)
```

## Question 30

```
def is_valid(string):
    ls = [] 
    openn = {'(' , '[' , '{'}
    close = {')' , ']' , '}'}

    for i in range(len(string)):

        if(string[i] in openn):
            ls.append(string[i])
        else:
            if(len(ls)==0):
                return False
            if(string[i]==')' and ls[len(ls)-1]!='('):
                return False
            if(string[i]==']' and ls[len(ls)-1]!='['):
                return False
            if(string[i]=='}' and ls[len(ls)-1]!='{'):
                return False
            else:
                ls.pop()

    return len(ls)==0

string = input("Enter your string :")


if(is_valid(string)):
    print("VALID")
else :
    print("NOT VALID")
            
```
