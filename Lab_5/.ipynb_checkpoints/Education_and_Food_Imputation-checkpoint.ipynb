## Part I — Indian Food Dataset (KNN Imputation)

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import KNNImputer
import numpy as np
import pandas as pd

# Load dataset
food = pd.read_csv('indian_food.csv')
print('Rows:', len(food))
print('prep_time counts (including -1):')
print(food['prep_time'].value_counts(dropna=False).head(10))

# Replace -1 with NaN
food['prep_time'] = pd.to_numeric(food['prep_time'], errors='coerce')
food.loc[food['prep_time'] == -1, 'prep_time'] = np.nan

# One-hot encode categorical columns
X_cat = food[['diet', 'state']].fillna('missing').astype(str)

# Compatibility with all sklearn versions
try:
    ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
except TypeError:
    ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_ohe = ohe.fit_transform(X_cat)

# Numeric column
prep = food[['prep_time']].to_numpy()

# Force both arrays to be 2D
if X_ohe.ndim == 1:
    X_ohe = X_ohe.reshape(-1, 1)
if prep.ndim == 1:
    prep = prep.reshape(-1, 1)

# Double-check shapes
print('Shapes before stacking:', X_ohe.shape, prep.shape)

# Stack safely
arr = np.hstack((X_ohe, prep))

# KNN Imputation
imputer = KNNImputer(n_neighbors=3)
imp = imputer.fit_transform(arr)

# Update the dataset
food['prep_time_imputed'] = imp[:, -1]

# Preview
print('\nImputed prep_time (first 10 rows):')
print(food[['name', 'diet', 'state', 'prep_time', 'prep_time_imputed']].head(10).to_string(index=False))
```

## Part II — PGI Dataset (Mean/Median/Mode Imputation)

```python
pgi = pd.read_csv('PGI.csv')
print('Overall column candidates:')
print([c for c in pgi.columns if 'overall' in c.lower()])
overall_col = [c for c in pgi.columns if 'overall' in c.lower()][0]
print('\nSummary before imputation:')
print(pgi[overall_col].describe())

pgi_mean = pgi.copy()
pgi_mean[overall_col] = pgi_mean[overall_col].fillna(pgi_mean[overall_col].mean())
pgi_median = pgi.copy()
pgi_median[overall_col] = pgi_median[overall_col].fillna(pgi_median[overall_col].median())
pgi_mode = pgi.copy()
mode_val = pgi_mode[overall_col].mode().iloc[0]
pgi_mode[overall_col] = pgi_mode[overall_col].fillna(mode_val)

print('\nAverages after imputations:')
print('Mean-imputed average:', pgi_mean[overall_col].mean())
print('Median-imputed average:', pgi_median[overall_col].mean())
print('Mode-imputed average:', pgi_mode[overall_col].mean())
```

## Part III — Punjab Girls Enrollment (Correlation Heatmaps)

```python
punjab = pd.read_csv('Punjab_Girls_Enrollment_Class11_12_By_Year.csv')
year_cols = [c for c in punjab.columns if c.strip().isdigit()]
if not year_cols:
    year_cols = [c for c in punjab.columns if any(str(y) in c for y in range(2009,2023))]
print('Year columns used:', year_cols)
punjab_years = punjab[year_cols].apply(pd.to_numeric, errors='coerce')
punjab_dropna = punjab_years.dropna(axis=0, how='any')
from sklearn.impute import KNNImputer
imputer2 = KNNImputer(n_neighbors=3)
punjab_imputed_arr = imputer2.fit_transform(punjab_years.fillna(punjab_years.mean()))
punjab_imputed = pd.DataFrame(punjab_imputed_arr, columns=punjab_years.columns)
corr_drop = punjab_dropna.corr()
corr_imputed = punjab_imputed.corr()
print('Shapes:', corr_drop.shape, corr_imputed.shape)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(corr_drop, cmap='RdBu_r', center=0)
plt.title('Correlation (dropna)')
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(corr_imputed, cmap='RdBu_r', center=0)
plt.title('Correlation (KNN imputed)')
plt.show()
# align and compute abs diff
common_idx = [c for c in corr_imputed.index if c in corr_drop.index]
if common_idx:
    corr_drop_aligned = corr_drop.reindex(index=common_idx, columns=common_idx)
else:
    corr_drop_aligned = corr_drop.reindex_like(corr_imputed)
abs_diff = (corr_drop_aligned - corr_imputed.loc[corr_drop_aligned.index, corr_drop_aligned.columns]).abs()
plt.figure(figsize=(8,6))
sns.heatmap(abs_diff, cmap='viridis')
plt.title('Absolute difference (dropna vs imputed)')
plt.show()
print('\nMax absolute difference:', abs_diff.max().max())
```
