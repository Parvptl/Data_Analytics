Q1

Set a random seed at the beginning: np.random.seed(42).

(a) Create the following NumPy arrays:
    1. np1: Integers from 1 to 100 (inclusive).
    2. np2: 100 samples from a normal distribution with mean = 170 and standard
        deviation = 10.
    3. np3: 100 samples from a uniform distribution between 60 and 90.
    4. np4: 100 random integers between 20,000 and 100,000 (inclusive).
    5. np5: Randomly assign one of the categories “HR”, “Sales”, “Tech”, or
        “Finance” to each element.
Use appropriate NumPy functions for each step.

```python
# Q 1.1
import numpy as np
np.random.seed(42)
np1 = np.arange(1,101,1)
print(np1)
```

```python
# Q 1.2
np2 = np.random.normal(loc=170,scale=10,size=100)
print(np2)
```

```python
# Q 1.3
np3 = np.random.uniform(low=60,high=90,size=100)
print(np3)
```

```python
# Q 1.4
np4 = np.random.randint(20000,100001,100)
print(np4)
```

```python
# Q 1.5
options_list = ["HR" , "Sales" , "Tech" , "Finance" ]
np5 = np.random.choice(options_list, size=100)
print(np5)
```

9. Temperature Conversion and Analysis (2)
(a) Convert the temperature columns (day temp, min temp, max temp) from Kelvin
to Celsius.
(b) Calculate the daily temperature range (max temp − min temp) for each day.
(c) Identify the city and date with the highest temperature range recorded.
(d) Compute the mean day temperature (in Celsius) for each city over the 7-day
period.

```python
# Q9 a
# To convert kelvin tempratures into celcius subtracting 273 from given kelvin temprature
dw['day_temp']-=273
dw['min_temp']-=273
dw['max_temp']-=273
```

```python
# Q9 b
dw['daily_temp_range'] = dw['max_temp']-dw['min_temp']
```

```python
# Q9 c
print(dw['city_name'][dw.idxmax()['daily_temp_range']])
print(dw['date'][dw.idxmax()['daily_temp_range']])
```

```python
# Q9 d
df_new = dw.groupby('city_name')['day_temp'].rolling(7).mean()
print(df_new)
```
