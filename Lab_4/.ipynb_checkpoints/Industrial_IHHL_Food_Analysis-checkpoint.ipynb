```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 120, 'font.size': 10})
```

## Part 1 – Industrial Characteristics Dataset

### Q1. Z-scores for Fixed Capital and Outlier Detection

```python
ind = pd.read_csv('Estimate_of_important_characteristics_by_State_2022 _2023.csv')
fixed_col = [c for c in ind.columns if 'fixed' in c.lower() and 'capital' in c.lower()][0]
ind[fixed_col] = pd.to_numeric(ind[fixed_col], errors='coerce')
ind['FixedCapital_z'] = (ind[fixed_col] - ind[fixed_col].mean()) / ind[fixed_col].std(ddof=0)
outliers = ind.loc[ind['FixedCapital_z'].abs() >= 2.5, [ind.columns[0], fixed_col, 'FixedCapital_z']]
print('Fixed Capital column used:', fixed_col)
print('\nOutliers (|z|>=2.5):')
print(outliers.to_string(index=False))
```

### Q2. 3-point Moving Average for Wages, Emoluments, and Net Income

```python
ind = pd.read_csv('Estimate_of_important_characteristics_by_State_2022 _2023.csv')
cols = ind.columns
wage_col = [c for c in cols if 'wage' in c.lower()][0]
emol_col = [c for c in cols if 'emol' in c.lower()][0]
net_col = [c for c in cols if 'net' in c.lower() and 'income' in c.lower()][0]
for col in [wage_col, emol_col, net_col]:
    ind[col] = pd.to_numeric(ind[col], errors='coerce')
    ind[col + '_MA3'] = ind[col].rolling(window=3, center=True, min_periods=1).mean()
print(ind[[ind.columns[0]] + [wage_col+'_MA3', emol_col+'_MA3', net_col+'_MA3']].head())
```

## Part 2 – IHHL Dataset

### Q3. Total Achieved IHHL per State

```python
ihhl = pd.read_csv('IHHL.csv')
state_col = [c for c in ihhl.columns if 'state' in c.lower()][0]
ihhl_col = [c for c in ihhl.columns if 'ihhl' in c.lower() and 'ach' in c.lower()][0]
ihhl_totals = ihhl.groupby(state_col)[ihhl_col].sum().reset_index().rename(columns={ihhl_col: 'Total_IHHLAch'})
print(ihhl_totals.head())
```

### Q4. Scatter Plot of Total IHHL per State (Outliers in Red)

```python
ihhl_totals['z'] = (ihhl_totals['Total_IHHLAch'] - ihhl_totals['Total_IHHLAch'].mean()) / ihhl_totals['Total_IHHLAch'].std(ddof=0)
colors = ['red' if abs(v) >= 2.5 else 'C0' for v in ihhl_totals['z']]
plt.figure(figsize=(12,6))
plt.scatter(ihhl_totals[state_col], ihhl_totals['Total_IHHLAch'], c=colors)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total IHHL Achieved')
plt.title('Total IHHL Achieved per State (Outliers in Red)')
plt.tight_layout()
plt.show()
```

## Part 3 – Indian Food Dataset

### Q5. Compute Average Total Time per State

```python
food = pd.read_csv('indian_food.csv')
food['prep_time'] = pd.to_numeric(food['prep_time'], errors='coerce').fillna(0)
food['cook_time'] = pd.to_numeric(food['cook_time'], errors='coerce').fillna(0)
food['total_time'] = food['prep_time'] + food['cook_time']
avg_total = food.groupby('state')['total_time'].mean().sort_values(ascending=False).reset_index()
print(avg_total.head(10))
```

### Q6. Pie Chart for Top 8 States by Average Cooking Time

```python
top8 = avg_total.head(8)
labels = top8['state']
sizes = top8['total_time']
plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 8 States by Average Total Cooking Time')
plt.tight_layout()
plt.show()
```
