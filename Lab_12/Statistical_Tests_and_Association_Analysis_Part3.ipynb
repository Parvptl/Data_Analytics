# Part III: Chi-Squared Test
OBJECTIVE: Chi-Square Test of Independence for analyzing how certain key features
of the dataset are related to social media usage
DATASET: “Students Social Media Addiction.csv”
DATASET DESCRIPTION: Students’ Distraction & Social Media Addiction Dataset;
As quoted from Kaggle, “A dataset exploring how social media affects students’ daily
life, study patterns; Includes data on how much students use social media daily, and
whether it affects their academic performance”. More details can be found here
PRELIMINARY:
Form two new categorical features as follows:
1. “Usage” and fill the values as “Overuse” or “Controlled ” based on the correspond-
ing feature “Avg Daily Usage Hours” ≥ or < median of the values, respectively.
2. “Sleep” and fill the values as “Sufficient” or “Insufficient” based on the corre-
sponding variable Sleep Hours Per Night ≥7 or <7, respectively.
TASK:
Generate the contingency table and apply the Chi-Square Test of Independence on each
of the following pairs formed by the respective features and “Usage”.

```python
import pandas as pd
```

```python
df = pd.read_csv("Students Social Media Addiction.csv")
```

```python
df
```

```python
median = df["Avg_Daily_Usage_Hours"].median()
```

```python
import numpy as np
```

```python
df["Usage"] = np.where(df['Avg_Daily_Usage_Hours']>median , "Overuse" , "Controlled")
```

```python
df["Sleep"] = np.where(df['Sleep_Hours_Per_Night']>=7 , "Sufficient" , "Insufficient")
```

```python
ct1 = pd.crosstab(index = df['Academic_Level'] , columns = df['Usage'])
display(ct1)
```

```python
ct2 = pd.crosstab(index = df['Gender'] , columns = df['Usage'])
display(ct2)
```

```python
ct3 = pd.crosstab(index = df['Sleep'] , columns = df['Usage'])
display(ct3)
```

```python
import numpy as np
from scipy.stats import chi2_contingency

res = chi2_contingency(ct1)
display(res.statistic)
display(res.pvalue)
display(res.dof)
display(res.expected_freq)
```

```python
res2 = chi2_contingency(ct2)
display(res2.statistic)
display(res2.pvalue)
display(res2.dof)
display(res2.expected_freq)
```

```python
res3 = chi2_contingency(ct3)
display(res3.statistic)
display(res3.pvalue)
display(res3.dof)
display(res3.expected_freq)
```

```python
thr1 = 3.841
thr2 = 5.991

data = {'Features': ['Academic Level', 'Gender', 'Sleep'],
        'Chi-Square Statistic': [res.statistic, res2.statistic, res3.statistic],
        'Threshould' : [5.991 , 3.841 ,3.841]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Is related to Usage"] = np.where(df['Chi-Square Statistic']>=df['Threshould'] , "YES" , "NO")
```
