```python
import pandas as pd
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
df
```

Let us say salary variability of Data Scientist are sigma1 and for Data Engineers are sigma 2

Null Hypothesis : sigma1 == sigma 2

Alternate Hypothesis : sigma1 !=sigma 2

```python
ds = df[df['job_title']=="Data Scientist"]
```

```python
de = df[df['job_title']=="Data Engineer"]
```

```python
from scipy import stats

statistic, pvalue = stats.levene(ds['salary_in_usd'],de['salary_in_usd'], center='mean')
```

```python
statistic
```

```python
pvalue
```

as p value is higher than our significance level we are fail to reject null hypothesis.

## Question 2 [2 Marks]: It is claimed that the average annual salary of all employees

listed in the dataset is $100,000. Using the dataset ‘ds salaries.csv’, test this claim using
a one-sample t-test with the ‘salary in usd’ column. Use a significance level of α = 0.05.
Tasks:
• State the null and alternative hypotheses. [0.5]
• Perform the one-sample t-test using scipy.stats.ttest 1samp. [0.5]

1. Null Hypothesis : avg_salary = $1000000
2. Alternate Hypothesis : avg_salary != $1000000

```python
statistic , pvalue = stats.ttest_1samp(df['salary_in_usd'], 100000, alternative='two-sided')
```

```python
statistic
```

```python
pvalue
```

as p value is less than our significance level we can reject null hypothesis
