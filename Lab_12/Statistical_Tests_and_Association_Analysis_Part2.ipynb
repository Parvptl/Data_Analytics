 ## Part A: Single Sample Z-Test (2.0 marks)
The company’s executive team claims that:
• The average Performance Score of employees is 3.0.
• The average Employee Satisfaction Score is 3.2.
Test both claims at a 5% significance level. Assume:
• Population standard deviation for Performance Score = 1.41
• Population standard deviation for Employee Satisfaction Score = 1.15
Tasks:
• For each claim, state the null and alternative hypotheses
• Calculate the sample mean, z-statistic.
• Make a decision (Reject/Fail to Reject H 0 ) for each test.
• Provide a summary table with results for both tests

```python
import numpy as np
```

1. Null Hypothesis : avg = 3
2. Alternate Hypothesis : avg !=3

```python
sample_mean = df['Performance_Score'].mean()
mu_0 = 3.0 # hypothesized population mean
sigma = 1.41 # known population standard deviation
n = len(df)
# Calculate Z-score
z_score = (sample_mean - mu_0) / (sigma / np.sqrt(n))
print(f"Z-score: {z_score}")
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(f"P-value: {p_value}")
```

as p value is higher than our significance level we are fail to reject null hypothesis

```python

# Create a summary
results = {
'Test Statistic (Z)': [z_score],
'P-value': [p_value],
'Significance Level': [0.05],
'Decision': ['Reject H₀' if p_value <= 0.05 else 'Fail to reject H₀']
}
results_df = pd.DataFrame(results)
print(results_df)
```

1. Null Hypothesis : avg = 3.2
2. Alternate Hypothesis : avg != 3.2

```python
sample_mean = df['Employee_Satisfaction_Score'].mean()
mu_0 = 3.2 # hypothesized population mean
sigma = 1.15 # known population standard deviation
n = len(df)
# Calculate Z-score
z_score = (sample_mean - mu_0) / (sigma / np.sqrt(n))
print(f"Z-score: {z_score}")
```

```python

p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(f"P-value: {p_value}")
```

as p value is higher than our significance level we are fail to reject null hypothesis

```python

# Create a summary
results = {
'Test Statistic (Z)': [z_score],
'P-value': [p_value],
'Significance Level': [0.05],
'Decision': ['Reject H₀' if p_value <= 0.05 else 'Fail to reject H₀']
}
results_df = pd.DataFrame(results)
print(results_df)
```

## Part B: Two-Sample Z-Test (2.0 marks)
The HR department wants to investigate:
1. Whether there is a significant difference in average Work Hours Per Week between
employees who have Resigned (True) vs those who have Not Resigned (False). Use
α = 0.05.
Page 22. Whether there is a significant difference in the average Monthly Salary between
Male and Female employees. Use α = 0.05.
Assume population standard deviations:
• Work Hours Per Week: 8.94 hours for both groups
• Monthly Salary: $1,370 for both groups
Tasks:
• For each comparison, state the null and alternative hypotheses.
• Calculate sample means for both groups, z-statistic.
• Make a decision (Reject/Fail to Reject H 0 ) for each test.
• Provide a summary table with results for both tests.

```python
res = df[df['Resigned']==True]
nres = df[df['Resigned']!=True]
```

1. Null Hypothesis : res_avg = nres_avg
2. Alternate Hypothesis : res_avg != nres_avg

```python
sample_mean1 = res['Work_Hours_Per_Week'].mean()
s1 = 8.94
n1= len(res)
```

```python
sample_mean2 = nres['Work_Hours_Per_Week'].mean()
s2 = 8.94
n2 = len(res)
```

```python
z = (sample_mean1-sample_mean2)/np.sqrt((s1**2/n1)+(s2**2/n2))
display(z)
```

```python
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value}")
```

as p value is higher than significance level we are fail to reject null hypothesis

```python

# Create a summary
results = {
'Test Statistic (Z)': [z],
'P-value': [p_value],
'Significance Level': [0.05],
'Decision': ['Reject H₀' if p_value <= 0.05 else 'Fail to reject H₀']
}
results_df = pd.DataFrame(results)
print(results_df)
```
