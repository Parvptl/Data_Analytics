## Q1. Synthetic Dataset with Equal-Density Clusters
Generate a synthetic dataset with three clusters of roughly the same density and
similar number of points per cluster. (Parameters: n samples=600, centers=3,
cluster std=0.6, random state=42)

# Tasks:
• Apply KMeans clustering on this dataset and report the SSE using the
built-in inertia attribute.
• Apply DBSCAN clustering and compute SSE.
• Plot the clusters obtained.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans , BisectingKMeans
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=600 , centers=3 , cluster_std= 0.6 , random_state=42)
```

```python
kmeans = KMeans(n_clusters=3 , random_state=42)

kmeans.fit(X)

y_kmeans = kmeans.predict(X)

centroids = kmeans.cluster_centers_

wcss = kmeans.inertia_

print(wcss)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.7, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
# Plot DBSCAN results
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
x1 = X[dbscan_labels==0]
x2 = X[dbscan_labels==1]
x3 = X[dbscan_labels==2]

x1_c = x1-x1.mean(axis=0)
x2_c = x2-x2.mean(axis=0)
x3_c = x3-x3.mean(axis=0)

x1_sq = [x[0]**2 + x[1]**2 for x in x1_c]
x2_sq = [x[0]**2 + x[1]**2 for x in x2_c]
x3_sq = [x[0]**2 + x[1]**2 for x in x3_c]

x1_sq = np.sqrt(x1_sq)
x2_sq = np.sqrt(x2_sq)
x3_sq = np.sqrt(x3_sq)

x_sum = sum(x2_sq) + sum(x2_sq) + sum(x2_sq)
x_sum
```

## Q2. Synthetic Dataset with Varying-Density Clusters
